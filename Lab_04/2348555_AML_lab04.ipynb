```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
# Load the dataset
dataset = pd.read_csv('/content/Program_4_Dataset.csv')
```

```python
dataset.describe()
```

* No null data
* Normal Distribution

```python
# Separate features and target variable
X = dataset['Temperature'].values.reshape(-1, 1)
y = dataset['Pressure'].values
```

* Target Variable = Pressure
* Predictor Variable = Temperature

```python
# Define polynomial regression degrees
degrees = [2, 3, 4, 5, 6, 7]
```

* Polynomial Degrees used are 2,3,4,5,6,7

```python
# Evaluate model performance for different train-test splits
for split_ratio in [0.66, 0.7, 0.8]:
    for degree in degrees:
        # Create polynomial features
        poly_features = PolynomialFeatures(degree=degree)
        X_poly = poly_features.fit_transform(X)

        # Split data into training and testing sets
        X_train, X_test, y_train, y_test = train_test_split(X_poly, y, test_size=1 - split_ratio, random_state=1)

        # Train the linear regression model
        model = LinearRegression()
        model.fit(X_train, y_train)

        # Predict pressure values for the test set
        y_pred = model.predict(X_test)

        # Calculate evaluation metrics
        mse = mean_squared_error(y_test, y_pred)
        mae = mean_absolute_error(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)

        # Print results
        print(f"Degree: {degree}, Split Ratio: {split_ratio}")
        print(f"R2 Score: {r2:.3f}")
        print(f"Mean Squared Error: {mse:.3f}")
        print(f"Mean Absolute Error: {mae:.3f}")
        print("----------------------------------")
```

Overall Performance

* The overall performance of the polynomial regression model is good, with R^2 scores ranging from 0.564 to 0.685 and mean squared error (MSE) values ranging from 22.249 to 30.592. This indicates that the model is able to capture a significant portion of the variability in the data and make reasonably accurate predictions.

Effect of Polynomial Degree

* The performance of the model generally increases with the polynomial degree, as indicated by the increasing R^2 scores and decreasing MSE values. This is because higher polynomial degrees allow the model to capture more complex relationships between the temperature and pressure variables. However, the improvement in performance diminishes with higher degrees, suggesting that overfitting may occur.

Effect of Train-Test Split Ratio

* The performance of the model is relatively consistent across the different train-test split ratios, indicating that the model is not overly sensitive to the size of the training and testing sets. This suggests that the model is robust and can generalize well to unseen data.

Detailed Observations

Here are some specific observations from the performance results:

* The best R^2 score (0.685) was achieved with a polynomial degree of 7 and a train-test split ratio of 0.66.
* The lowest MSE value (22.249) was achieved with a polynomial degree of 7 and a train-test split ratio of 0.66.
* The overall trend suggests that a polynomial degree of 6 or 7 provides a good balance between model fit and generalization.
Recommendations

Based on the analysis of the performance results, the following recommendations can be made:

* For the best overall performance, use a polynomial degree of 6 or 7.
Use a train-test split ratio of 0.66 to balance the size of the training and testing sets.
* Be aware of the potential for overfitting with higher polynomial degrees.
